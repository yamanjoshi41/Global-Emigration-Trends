```python
# Total emigration by year
emigration_by_year = df.groupby('Year')['Total number of emigrants'].sum()
emigration_by_year
```

```python
# Emigration Trend Over Time
plt.figure(figsize=(10, 6))
plt.plot(emigration_by_year, marker='o', color='b', linestyle='--')
plt.title('Global Emigration Trend Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Number of Emigrants', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

```python
top_countries = df.groupby('Entity')['Total number of emigrants'].sum().sort_values(ascending=False).head(10)
top_countries
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="mako")
plt.title('Top 10 Countries with Highest Emigration', fontsize=16)
plt.xlabel('Total Number of Emigrants', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', alpha=0.5)
plt.show()
```
